```python
class InsufficientFundsError(Exception):
    """Custom exception for when withdrawal amount exceeds balance."""
    pass

class BankAccount:
    def __init__(self, owner, matric_no, initial_balance=0.0):
        if initial_balance < 0:
            raise ValueError("Initial balance cannot be negative.")

        self.owner = owner
        self.matric_no = matric_no
        self.__balance = float(initial_balance) # Private attribute
        self.transaction_fee = 0.50

    @property
    def balance(self):
        return self.__balance

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit amount must be positive.")
        self.__balance += amount
        print(f"Deposited: ${amount:.2f}. New Balance: ${self.__balance:.2f}")

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive.")

        total_deduction = amount + self.transaction_fee
        if total_deduction > self.__balance:
            raise InsufficientFundsError(f"Insufficient funds for withdrawal of ${amount:.2f} + fee.")

        self.__balance -= total_deduction
        print(f"Withdrew: ${amount:.2f} (Fee: ${self.transaction_fee:.2f}). New Balance: ${self.__balance:.2f}")

    @classmethod
    def from_dict(cls, data):
        return cls(data['owner'], data['matric_no'], data.get('balance', 0.0))

    def __str__(self):
        return f"Account({self.owner}, {self.matric_no}, Balance: ${self.__balance:.2f})"

    def __repr__(self):
        return self.__str__()

def run_tests():
    print("--- Starting BankAccount Tests ---")

    # 1. Test Initialization
    account = BankAccount("Oyebode Omobolaji Ebunoluwa", "CPE/2023/1101", 100.0)
    print(f"Test 1: Initialization Successful -> {account}")

    # 2. Test Deposit
    try:
        account.deposit(50.0)
        print("Test 2: Deposit Passed.")
    except Exception as e:
        print(f"Test 2: Failed -> {e}")

    # 3. Test Withdrawal
    try:
        account.withdraw(20.0)
        print("Test 3: Withdrawal Passed.")
    except Exception as e:
        print(f"Test 3: Failed -> {e}")

    # 4. Test Insufficient Funds
    try:
        account.withdraw(500.0)
    except InsufficientFundsError:
        print("Test 4: InsufficientFundsError caught correctly.")

    # 5. Test from_dict
    data = {'owner': 'Oyebode Omobolaji Ebunoluwa', 'matric_no': 'CPE/2023/1101', 'balance': 500}
    new_acc = BankAccount.from_dict(data)
    print(f"Test 5: from_dict created -> {new_acc}")

    print("--- All Tests Completed ---")

if __name__ == "__main__":
    run_tests()
```
